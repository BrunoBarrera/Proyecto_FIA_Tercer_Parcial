# Creacion del dataset
En esta sección, adaptaremos los datos sin limpiar a un formato que podamos manipular a lo largo del proceso

```python
import pandas as pd
import pickle
import os
```

```python
path = "dataset"
classifications = ['Cerebro','Descubrimientos_recientes','Historia_y_areas_de_la_medicina','Identificación_de_causas_de_enfermedades','Sistemas_del_ser_humano','Tratamientos_procedimientos_y_medicamentos']
df = pd.DataFrame(columns=['file_name','content','category'])
content = []
classificationCol = []
categoryID = []
filename = []
for classification in classifications:
    for root, dirs, files in os.walk(os.path.join(path, '{}'.format(classification)), topdown=False):
        for name in files:
            filePath = os.path.join(root, name)
            fileContent = open(filePath,mode='r',encoding="UTF-8", errors='ignore').read()
            content.append(fileContent)
            classificationCol.append(classification)
            categoryID.append(classifications.index(classification) )
            filename.append(name)

df["content"] = content
df["category"] = classificationCol
df["file_name"] = filename
df
```

```python
# Exportamos
with open('dataframes/dirty_data.pickle', 'wb') as output:
    pickle.dump(df, output)
with open('dataframes/labels.pickle', 'wb') as output:
    pickle.dump(categoryID, output)

```
